```python
bank_data = pd.read_csv('Portuguese_banking.csv')
df_bank = bank_data.copy()
df_bank.head()
```

```python
import matplotlib.pyplot as plt  
%matplotlib inline

figure = plt.figure(figsize =(5, 7)) 
plt.boxplot(df_bank['age'])
plt.show()   #Outlierler var
```

```python
#Preparing Data for modelling 
df_new = pd.get_dummies(data = df_bank, 
                        columns=['housing','job','marital','education', 'default', 'loan', 'contact', 
                                 'month','day_of_week', 'poutcome', 'y' ], drop_first=True)
df_new.columns.values[-1]='TARGET'
df_new
```

```python
from sklearn.model_selection import train_test_split 
from sklearn import metrics

X_train, X_test, y_train, y_test = train_test_split(df_new.drop('TARGET',axis=1), 
           df_new['TARGET'], test_size=0.30, 
            random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression  
logModel = LogisticRegression(max_iter = 100000)
logModel.fit(X_train, y_train)
predictions = logModel.predict(X_test)
predictions
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
